# Data Visualization with Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print('libraries loaded!')
```

```python
df = pd.read_csv('2019.csv')
print('dataset loaded')
print('shape:', df.shape)
df.head()
```

```python
print('column names:')
print(df.columns.tolist())
print()
print('missing values:')
print(df.isnull().sum())
```

```python
print('basic statistics:')
df.describe()
```

```python
top10 = df.head(10)

plt.figure(figsize=(10, 5))
plt.barh(top10['Country or region'], top10['Score'], color='steelblue')
plt.title('Top 10 Happiest countries in 2019')
plt.xlabel('Happiness Score')
plt.ylabel('country')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
bottom10 = df.tail(10)

plt.figure(figsize=(10, 5))
plt.barh(bottom10['Country or region'], bottom10['Score'], color='red')
plt.title('Bottom 10 least Happy Countries in 2019')
plt.xlabel('Happiness Score')
plt.ylabel('country')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['GDP per capita'], df['Score'], color='green', alpha=0.6)
plt.title('GDP per capita vs Happiness score')
plt.xlabel('GDP per Capita')
plt.ylabel('happiness Score')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Social support'], df['Score'], color='purple', alpha=0.6)
plt.title('Social support vs Happiness score')
plt.xlabel('Social Support')
plt.ylabel('Happiness score')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Score'], bins=20, color='orange')
plt.title('Distribution of Happiness Scores')
plt.xlabel('happiness score')
plt.ylabel('Number of countries')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr = df[['Score', 'GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('correlation heatmap of happiness factors')
plt.tight_layout()
plt.show()
```

```python
print('findings from the visualization:')
print()
print('1. Finland is the happiest country in 2019')
print('2. GDP per capita has strong positive correlation with happiness')
print('3. Social support is also a major factor for happiness')
print('4. Most countries have happiness score between 4 and 6')
print('5. Countries with low GDP tend to have low happiness scores')
print('6. Freedom to make life choices positively impacts happiness')
print('7. Generosity has the weakest correlation with happiness score')
print('8. Healthy life expectancy is strongly linked to happiness')
print()
print('Happiest country:', df.loc[df['Score'].idxmax(), 'Country or region'])
print('least happy country:', df.loc[df['Score'].idxmin(), 'Country or region'])
print('Average happiness score:', round(df['Score'].mean(), 2))
```
