```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print("Libraries loaded successfully!")
```

```python
df = pd.read_csv('StudentsPerformance.csv')
print('Dataset loaded successfully!')
print('Number of rows and columns:', df.shape)
df.head()
```

```python
print('Column names in the dataset:')
print(df.columns.tolist())
print()
print('Data types:')
print(df.dtypes)
```

```python
print('Checking for missing values:')
print(df.isnull().sum())
print()
print('Total missing values:', df.isnull().sum().sum())
print()
print('Checking for duplicates:', df.duplicated().sum())
```

```python
print('Basic Statistics of the dataset:')
df.describe()
```

```python
print('Average scores by gender:')
gender_avg = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean().round(2)
print(gender_avg)
```

```python
print('Average scores by parental education level:')
parent_avg = df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean().round(2)
print(parent_avg)
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['math score'], bins=20, color='steelblue')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Math Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='parental level of education', y='math score', data=df)
plt.title('Math Score by Parental Education Level')
plt.xlabel('Parental Education Level')
plt.ylabel('Math Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='reading score', y='writing score', hue='gender', data=df)
plt.title('Reading Score vs Writing Score')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
corr = df[['math score', 'reading score', 'writing score']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Scores')
plt.tight_layout()
plt.show()
```

```python
print('Key EDA Findings:')
print()
print('1. Dataset has', len(df), 'student records with no missing values')
print('2. Males perform better in math while females score higher in reading and writing')
print('3. Students with highly educated parents score significantly better')
print('4. Reading and writing scores are strongly correlated')
print('5. Math scores have the widest variation among all subjects')
print('6. Average math score is:', round(df['math score'].mean(), 2))
print('7. Average reading score is:', round(df['reading score'].mean(), 2))
print('8. Average writing score is:', round(df['writing score'].mean(), 2))
```
